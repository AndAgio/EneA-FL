```python
import os
import pandas as pd
import numpy as np

BASE_DIR = "../"
DEVICES = ["cpu/", "gpu/"]
IOT_MODELS = ["jetson_nano/", "raspberrypi/", "jetson_orin/", "jetson_xavier/"]
```

```python
series_sent140 = {}
series_femnist = {}
series_nbaiot = {} 
for iot_model in IOT_MODELS:
    for device in DEVICES:
        path = os.path.join(BASE_DIR, iot_model, device, "split_dfs/")
        if not os.path.exists(path):
            continue

        key_device = "{}_{}".format(iot_model, device).replace("/", "")
        if key_device not in series_sent140:
            series_sent140[key_device] = []
        if key_device not in series_femnist:
            series_femnist[key_device] = []
        if key_device not in series_nbaiot:
            series_nbaiot[key_device] = []

        epoch = "epoch"

        df0_sent140 = list(pd.read_csv(os.path.join(path, "0_big_sent140_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df1_sent140 = list(pd.read_csv(os.path.join(path, "1_big_sent140_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df2_sent140 = list(pd.read_csv(os.path.join(path, "2_big_sent140_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df3_sent140 = list(pd.read_csv(os.path.join(path, "3_big_sent140_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df4_sent140 = list(pd.read_csv(os.path.join(path, "4_big_sent140_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df_sent140 = df0_sent140 + df1_sent140 + df2_sent140 + df3_sent140 + df4_sent140

        df0_femnist = list(pd.read_csv(os.path.join(path, "0_big_femnist_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df1_femnist = list(pd.read_csv(os.path.join(path, "1_big_femnist_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df2_femnist = list(pd.read_csv(os.path.join(path, "2_big_femnist_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df3_femnist = list(pd.read_csv(os.path.join(path, "3_big_femnist_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df4_femnist = list(pd.read_csv(os.path.join(path, "4_big_femnist_5{}.csv".format(epoch)))["energon_total_in_power_mW"])
        df_femnist = df0_femnist + df1_femnist + df2_femnist + df3_femnist + df4_femnist

        df0_nbaiot = list(pd.read_csv(os.path.join(path, "0_big_nbaiot_5epoch.csv".format(epoch)))["energon_total_in_power_mW"])
        df1_nbaiot = list(pd.read_csv(os.path.join(path, "1_big_nbaiot_5epoch.csv".format(epoch)))["energon_total_in_power_mW"])
        df2_nbaiot = list(pd.read_csv(os.path.join(path, "2_big_nbaiot_5epoch.csv".format(epoch)))["energon_total_in_power_mW"])
        df3_nbaiot = list(pd.read_csv(os.path.join(path, "3_big_nbaiot_5epoch.csv".format(epoch)))["energon_total_in_power_mW"])
        df4_nbaiot = list(pd.read_csv(os.path.join(path, "4_big_nbaiot_5epoch.csv".format(epoch)))["energon_total_in_power_mW"])
        df_nbaiot = df0_nbaiot + df1_nbaiot + df2_nbaiot + df3_nbaiot + df4_nbaiot

        dots_sent140_y = [df0_sent140[0], df1_sent140[0], df2_sent140[0], df3_sent140[0], df4_sent140[0], df4_sent140[-1]]
        dots_sent140_x = [0, len(df0_sent140)]
        dots_sent140_x.append(len(df0_sent140) + len(df1_sent140))
        dots_sent140_x.append(len(df0_sent140) + len(df1_sent140) + len(df2_sent140))
        dots_sent140_x.append(len(df0_sent140) + len(df1_sent140) + len(df2_sent140) + len(df3_sent140))
        dots_sent140_x.append(len(df0_sent140) + len(df1_sent140) + len(df2_sent140) + len(df3_sent140) + len(df4_sent140))

        dots_femnist_y = [df0_femnist[0], df1_femnist[0], df2_femnist[0], df3_femnist[0], df4_femnist[0], df4_femnist[-1]]
        dots_femnist_x = [0, len(df0_femnist)]
        dots_femnist_x.append(len(df0_femnist) + len(df1_femnist))
        dots_femnist_x.append(len(df0_femnist) + len(df1_femnist) + len(df2_femnist))
        dots_femnist_x.append(len(df0_femnist) + len(df1_femnist) + len(df2_femnist) + len(df3_femnist))
        dots_femnist_x.append(len(df0_femnist) + len(df1_femnist) + len(df2_femnist) + len(df3_femnist) + len(df4_femnist))

        dots_nbaiot_y = [df0_nbaiot[0], df1_nbaiot[0], df2_nbaiot[0], df3_nbaiot[0], df4_nbaiot[0], df4_nbaiot[-1]]
        dots_nbaiot_x = [0, len(df0_nbaiot)]
        dots_nbaiot_x.append(len(df0_nbaiot) + len(df1_nbaiot))
        dots_nbaiot_x.append(len(df0_nbaiot) + len(df1_nbaiot) + len(df2_nbaiot))
        dots_nbaiot_x.append(len(df0_nbaiot) + len(df1_nbaiot) + len(df2_nbaiot) + len(df3_nbaiot))
        dots_nbaiot_x.append(len(df0_nbaiot) + len(df1_nbaiot) + len(df2_nbaiot) + len(df3_nbaiot) + len(df4_nbaiot))

        series_sent140[key_device].append((df_sent140, list(range(len(df_sent140)))))
        series_sent140[key_device].append((dots_sent140_y, dots_sent140_x))

        series_femnist[key_device].append((df_femnist, list(range(len(df_femnist)))))
        series_femnist[key_device].append((dots_femnist_y, dots_femnist_x))

        series_nbaiot[key_device].append((df_nbaiot, list(range(len(df_nbaiot)))))
        series_nbaiot[key_device].append((dots_nbaiot_y, dots_nbaiot_x))
  
```

```python
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
from matplotlib import rcParams

for font in font_manager.findSystemFonts("../../grafiche"):
    font_manager.fontManager.addfont(font)
rcParams['font.family'] = 'Palatino'
plt.figure().set_figwidth(15)

colors = ["lightblue", "#fad6a5", "#FF7276", "pink", "purple", "#E7D7AF", "#90EE90", "gray"]
dark_colors =  ["blue", "darkorange", "#B30000", "deeppink","darkviolet", "darkgoldenrod", "green",  "dimgray"]
markers = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X',]
linestyles = ['-.', ':', 'solid', '--', 'dashed', 'dashdot', 'dotted', '-', '--', '-.', ':',]
```

```python
pow_value = 3
def custom_transform(x):
    r = np.power(x, 1/pow_value)
    # check if np r array containeys nan
    if np.isnan(r).any():
        # repplace nan with 0
        r = np.nan_to_num(r)
    return r

def custom_inverse_transform(x):
    return np.power(x, pow_value)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))  # Adjust the figsize as needed
# make subplots share close to each other
fig.subplots_adjust(wspace=0.1)

for counter, k in enumerate(series_sent140.keys()):

    y_values = series_sent140[k][0][0]
    x_values = series_sent140[k][0][1]

    highlight_y = series_sent140[k][1][0]
    highlight_x = series_sent140[k][1][1]

    label = k.replace("_", " ")
    label = k.replace("_", " ")
    if "raspberrypi" in k:
        label = "raspberryPi4 cpu"

    axs[0].plot(x_values, y_values, color=colors[counter])
    axs[0].plot(highlight_x, highlight_y, color=dark_colors[counter], zorder=5, marker=markers[counter], markersize=5, label=label)

axs[0].set_xlabel('Time [seconds]')
axs[0].set_ylabel('Power [mW]')

axs[0].grid(True)

axs[0].set_yscale("functionlog", functions=[lambda x: x * 1, lambda x: x / 1])
axs[0].set_yticks([3000, 4000, 7000, 15000, 50000])
axs[0].set_yticklabels([3000, 4000, 7000, 15000, 50000])
axs[0].set_ylim([2000, 55000])


axs[0].set_xscale("function", functions=[custom_transform, custom_inverse_transform])
axs[0].set_xticks([0, 100, 500, 1000, 3000])
axs[0].set_xticklabels([0, 100, 500, 1000, 3000])

for counter, k in enumerate(series_femnist.keys()):
    # Create the line plots

    # if "orin" in k:
    #     counter -= 1
    #     continue

    y_values = series_femnist[k][0][0]
    x_values = series_femnist[k][0][1]

    highlight_y = series_femnist[k][1][0]
    highlight_x = series_femnist[k][1][1]

    label = k.replace("_", " ")
    if "raspberrypi" in k:
        label = "raspberryPi4 cpu"

    axs[1].plot(x_values, y_values, color=colors[counter])
    axs[1].plot(highlight_x, highlight_y, color=dark_colors[counter], zorder=5, marker=markers[counter], markersize=5, label=label)


# Add labels and title
axs[1].set_xlabel('Time [seconds]')
# hide y labels

# Add legend and grid
axs[1].grid(True)

axs[1].set_yscale("functionlog", functions=[lambda x: x * 10, lambda x: x / 10])
axs[1].set_yticks([3000, 4000, 7000, 15000, 50000])
axs[1].set_yticklabels([])
axs[1].set_ylim([2000, 55000])

axs[1].set_xscale("function", functions=[custom_transform, custom_inverse_transform])
axs[1].set_xticks([0, 100, 400, 1000, 3000, 6000, 10000])
axs[1].set_xticklabels([0, 100, 400, 1000, 3000, 6000, 10000])

for counter, k in enumerate(series_nbaiot.keys()):

    y_values = series_nbaiot[k][0][0]
    x_values = series_nbaiot[k][0][1]

    highlight_y = series_nbaiot[k][1][0]
    highlight_x = series_nbaiot[k][1][1]

    label = k.replace("_", " ")
    if "raspberrypi" in k:
        label = "raspberryPi4 cpu"

    axs[2].plot(x_values, y_values, color=colors[counter])
    axs[2].plot(highlight_x, highlight_y, color=dark_colors[counter], zorder=5, marker=markers[counter], markersize=5, label=label)
    # axs[2].scatter(highlight_x, highlight_y, linestyle=linestyles[counter], color=dark_colors[counter], marker=markers[counter], zorder=5, s=15)


# Add labels and title
axs[2].set_xlabel('Time [seconds]')
# axs[2].title('Power consuption over 5 epochs on MNIST dataset')

# Add legend and grid
axs[2].grid(True)

axs[2].set_yscale("functionlog", functions=[lambda x: x * 10, lambda x: x / 10])
axs[2].set_yticks([3000, 4000, 7000, 15000, 50000])
axs[2].set_yticklabels([])
# set y limits
axs[2].set_ylim([2000, 55000])

axs[2].set_xscale("function", functions=[custom_transform, custom_inverse_transform])
axs[2].set_xticks([0, 400, 1000, 2000, 3000, 5000, 6500])
axs[2].set_xticklabels([0, 400, 1000, 2000, 3000, 5000, 6500])

axs[0].text(0.5, -0.2, '(a)', ha='center', va='center', transform=axs[0].transAxes)
axs[1].text(0.5, -0.2, '(b)', ha='center', va='center', transform=axs[1].transAxes)
axs[2].text(0.5, -0.2, '(c)', ha='center', va='center', transform=axs[2].transAxes)
axs[2].legend(loc='upper right', bbox_to_anchor=(1, 0.95))

# save figure to pdf
plt.savefig("./pdfs/energies.pdf", bbox_inches='tight')

```
