```python
import os
import pandas as pd
import numpy as np

BASE_DIR = "./"
```

### SENT 140


```python
path = os.path.join(BASE_DIR, "sent140", "os_execution.csv")
os_execution = pd.read_csv(path)
os_execution_energy = list(os_execution[os_execution["timestamp"] > 1690083774.060438]["energon_total_in_power_mW"])
os_execution_timestamp = list(os_execution[os_execution["timestamp"] > 1690083774.060438]["timestamp"] - 1690083774.060438)

path = os.path.join(BASE_DIR, "sent140", "container_execution.csv")
container_execution = pd.read_csv(path)
container_execution_energy = list(container_execution[container_execution["timestamp"] > 1690084235.8575392]["energon_total_in_power_mW"])
container_execution_timestamp = list(container_execution[container_execution["timestamp"] > 1690084235.8575392]["timestamp"] - 1690084235.8575392 - 10)
```

```python
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
from matplotlib import rcParams

for font in font_manager.findSystemFonts("../../grafiche"):
    font_manager.fontManager.addfont(font)
rcParams['font.family'] = 'Palatino'
plt.figure().set_figwidth(15)

colors = ["lightblue", "#fad6a5", "#FF7276", "pink", "purple", "#E7D7AF", "#90EE90", "gray"]
dark_colors =  ["blue", "darkorange", "#B30000", "deeppink","darkviolet", "darkgoldenrod", "green",  "dimgray"]
markers = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X',]
linestyles = ['-.', ':', 'solid', '--', 'dashed', 'dashdot', 'dotted', '-', '--', '-.', ':',]
```

```python
plt.plot(os_execution_timestamp, os_execution_energy, color=colors[7], label="OS", marker=markers[0])
plt.plot(container_execution_timestamp, container_execution_energy, color=colors[2], label="Container", marker=markers[1])


# Add labels and title
plt.xlabel('Time [seconds]')
plt.ylabel('Energy [mW]')
plt.title('Energy consuption and training time over on Sent140 dataset')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
