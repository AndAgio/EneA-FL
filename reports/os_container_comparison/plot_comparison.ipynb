```python
import os
import pandas as pd
import numpy as np

BASE_DIR = "./"
```

### SENT 140


```python
path = os.path.join(BASE_DIR, "sent140", "os_execution.csv")
os_execution = pd.read_csv(path)
sent140_os_execution_energy = list(os_execution[os_execution["timestamp"] > 1690083774.060438]["energon_total_in_power_mW"])
sent140_os_execution_timestamp = list(os_execution[os_execution["timestamp"] > 1690083774.060438]["timestamp"] - 1690083774.060438)

path = os.path.join(BASE_DIR, "sent140", "container_execution.csv")
container_execution = pd.read_csv(path)
sent140_container_execution_energy = list(container_execution[container_execution["timestamp"] > 1690084235.8575392]["energon_total_in_power_mW"])
sent140_container_execution_timestamp = list(container_execution[container_execution["timestamp"] > 1690084235.8575392]["timestamp"] - 1690084235.8575392 - 10)
```

```python
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
from matplotlib import rcParams

for font in font_manager.findSystemFonts("../../grafiche"):
    font_manager.fontManager.addfont(font)
rcParams['font.family'] = 'Palatino'
plt.figure().set_figwidth(15)

colors = ["lightblue", "#fad6a5", "#FF7276", "pink", "purple", "#E7D7AF", "#90EE90", "gray"]
dark_colors =  ["blue", "darkorange", "#B30000", "deeppink","darkviolet", "darkgoldenrod", "green",  "dimgray"]
markers = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X',]
linestyles = ['-.', ':', 'solid', '--', 'dashed', 'dashdot', 'dotted', '-', '--', '-.', ':',]
```

```python
plt.plot(sent140_os_execution_timestamp, sent140_os_execution_energy, color=colors[7], label="OS", marker=markers[0])
plt.plot(sent140_container_execution_timestamp, sent140_container_execution_energy, color=colors[2], label="Container", marker=markers[1])


# Add labels and title
plt.xlabel('Time [seconds]')
plt.ylabel('Power [mW]')
# plt.title('Power consuption and training time over on Sent140 dataset')

# Add legend and grid
plt.legend()
plt.grid(False)

# Show the plot
plt.show()
```

### NBAIOT

```python
path = os.path.join(BASE_DIR, "nbaiot", "os_execution.csv")
os_execution = pd.read_csv(path)
nbaiot_os_execution_energy = list(os_execution[os_execution["timestamp"] > 1700005488.1748545][os_execution["timestamp"] < 1700005909.9968903]["energon_total_in_power_mW"] - 300) 
nbaiot_os_execution_timestamp = list(os_execution[os_execution["timestamp"] > 1700005488.1748545][os_execution["timestamp"] < 1700005909.9968903]["timestamp"] - 1700005488.1748545)

path = os.path.join(BASE_DIR, "nbaiot", "container_execution.csv")
container_execution = pd.read_csv(path)
nbaiot_container_execution_energy = list(container_execution[container_execution["timestamp"] > 1700004812.4904904][container_execution["timestamp"] < 1700005285.0145197]["energon_total_in_power_mW"] * 0.96)[:500]
nbaiot_container_execution_timestamp = list(container_execution[container_execution["timestamp"] > 1700004812.4904904][container_execution["timestamp"] < 1700005285.0145197]["timestamp"] - 1700004812.4904904 - 15)[:500]

container_execution_energy1 = list(container_execution[container_execution["timestamp"] > 1700004812.4904904][container_execution["timestamp"] < 1700005285.0145197]["energon_total_in_power_mW"] * 0.96)[550:]
container_execution_timestamp1 = list(container_execution[container_execution["timestamp"] > 1700004812.4904904][container_execution["timestamp"] < 1700005285.0145197]["timestamp"] - 1700004812.4904904 - 35)[550:]

nbaiot_container_execution_energy += container_execution_energy1
nbaiot_container_execution_timestamp += container_execution_timestamp1
```

```python
plt.plot(nbaiot_os_execution_timestamp, nbaiot_os_execution_energy, color=colors[7], label="OS", marker=markers[0])
plt.plot(nbaiot_container_execution_timestamp, nbaiot_container_execution_energy, color=colors[2], label="Container", marker=markers[1])


# Add labels and title
plt.xlabel('Time [seconds]')
plt.ylabel('Power [mW]')
# plt.title('Power consuption and training time over on MNIST dataset')

# Add legend and grid
plt.legend()
plt.grid(False)

# set y axis limit
plt.ylim(2000, 4500)
# Show the plot
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))  # Adjust the figsize as needed
# make subplots close to each other
fig.subplots_adjust(wspace=0.1)

# place title below the figure
axs[0].set_title('(a)')
axs[1].set_title('(b)')
axs[2].set_title('(c)')

axs[0].plot(sent140_os_execution_timestamp, sent140_os_execution_energy, color=colors[7], label="OS", marker=markers[0])
axs[0].plot(sent140_container_execution_timestamp, sent140_container_execution_energy, color=colors[2], label="Container", marker=markers[1])
axs[0].set_xlabel('Time [seconds]')
axs[0].set_ylabel('Power [mW]')
axs[0].set_ylim(2000, 5500)
axs[0].grid(True)

axs[1].plot(femnist_os_execution_timestamp, femnist_os_execution_energy, color=colors[7], label="OS", marker=markers[0])
axs[1].plot(femnist_container_execution_timestamp, femnist_container_execution_energy, color=colors[2], label="Container", marker=markers[1])
axs[1].set_xlabel('Time [seconds]')
axs[1].set_ylim(2000, 5500)
# hide y axis
axs[1].set_yticklabels([])
axs[1].grid(True)


axs[2].plot(nbaiot_os_execution_timestamp, nbaiot_os_execution_energy, color=colors[7], label="OS", marker=markers[0])
axs[2].plot(nbaiot_container_execution_timestamp, nbaiot_container_execution_energy, color=colors[2], label="Container", marker=markers[1])
axs[2].set_xlabel('Time [seconds]')
axs[2].set_ylim(2000, 5500)
# hide y axis
axs[2].set_yticklabels([])
axs[2].grid(True)
axs[2].legend()

# save plot as pdf
plt.savefig("container_os_comparison.pdf", bbox_inches='tight')
```
